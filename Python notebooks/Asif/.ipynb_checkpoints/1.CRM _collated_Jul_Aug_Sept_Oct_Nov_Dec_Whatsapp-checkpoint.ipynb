```python
data = df_queries_WA[['Number','User ID','Time1','User Query','Intent Recognized','Bot Response']]
```

```python
data['month'] = data['Time1'].dt.month
```

```python
data.head()
data.shape
```

```python
data.dtypes
```

```python
data['User ID'] = data['User ID'].astype(str)
data['Date_str'] = data['Date_1'].astype(str)
data['User Query'] = data['User Query'].astype(str)
data['Intent Recognized'] = data['Intent Recognized'].astype(str)
data['Bot Response'] = data['Bot Response'].astype(str)
```

```python
data.dtypes
```

# Jul_month_Base_Data

```python
#filter Month - Jul
data_7 = data[data['month']== 7]
data_7.shape

#Sort Data 
data_Jul = data_7.sort_values(by=['User ID','Time1','Number'], ascending=[1,1,0])
data_Jul = data_Jul[['User ID','Time1','User Query','Intent Recognized','Bot Response','Date_str','month']]
data_Jul.head()

#Define Sessions

data_Jul['Time'] = pd.to_datetime(data_Jul['Time1'])
data_Jul['diffs'] = data_Jul.groupby('User ID')['Time1'].diff()
data_Jul['break'] = data_Jul['diffs'] >= pd.Timedelta('10 min')

data_Jul['sessions'] = data_Jul.groupby('User ID')['break'].cumsum().apply(lambda x: 'session_{}'.format(x+1))
data_Jul.head(2)

#For Joining convert Variable to str

data_Jul['id_str']=data_Jul['User ID'].astype('str')
data_Jul['Date_str2']=data_Jul['Date_str'].astype('str')
data_Jul['Session_str']=data_Jul['sessions'].astype('str')

#Concat all three variables for joining
data_Jul['concat'] = data_Jul['id_str'] + data_Jul['Date_str2']+ data_Jul['Session_str']

# Create Unique Base for Joining
data_Base_Jul=data_Jul[['User ID','Date_str2','concat','month','sessions']]
data_Base_Jul.drop_duplicates(keep='first', inplace=True)
data_Base_Jul.head()
data_Base_Jul.shape

#User Query Transpose
data_Jul_User = data_Jul.groupby(['User ID','sessions','month','concat'])['User Query'].apply(lambda x: ' > '.join(x)).reset_index()
data_Jul_User['User Query'] = data_Jul_User['User Query'].str.lower()
data_Jul_User.head()

#Bot Response Transpose
data_Jul_Bot = data_Jul.groupby(['User ID','sessions','month','concat'])['Bot Response'].apply(lambda x: ' > '.join(x)).reset_index()
data_Jul_Bot['Bot Response'] = data_Jul_Bot['Bot Response'].str.lower()
data_Jul_Bot.head()

#Intent Response Transpose
data_Jul_intent = data_Jul.groupby(['User ID','sessions','month','concat'])['Intent Recognized'].apply(lambda x: ' > '.join(x)).reset_index()
data_Jul_intent['Intent Recognized'] = data_Jul_intent['Intent Recognized'].str.lower()
data_Jul_intent.head()

#Merge1
data_Base_Jul['key1'] = data_Base_Jul.concat.str.strip()
data_Jul_User['key1'] = data_Jul_User.concat.str.strip()
df_merge1 = pd.merge(data_Base_Jul,data_Jul_User, how='left', left_on=['key1'], right_on=['key1'])
df_merge1.head()
df_merge1.columns
df_merge1.drop(['key1','User ID_y', 'sessions_y', 'month_y', 'concat_y'],axis=1, inplace=True)
df_merge1.head(2)


#Merge2
df_merge1['key1'] = df_merge1.concat_x.str.strip()
data_Jul_Bot['key1'] = data_Jul_Bot.concat.str.strip()
df_merge2 = pd.merge(df_merge1,data_Jul_Bot, how='left', left_on=['key1'], right_on=['key1'])
df_merge2.head()
df_merge2.drop(['User ID_x', 'concat_x', 'month_x', 'sessions_x','key1'],axis=1, inplace=True)
df_merge2.head(2)

#Merge3
df_merge2['key1'] = df_merge2.concat.str.strip()
data_Jul_intent['key1'] = data_Jul_intent.concat.str.strip()
df_merge3 = pd.merge(df_merge2,data_Jul_intent, how='left', left_on=['key1'], right_on=['key1'])
df_merge3.head()
df_merge3.drop(['User ID_x', 'sessions_x', 'month_x', 'concat_x','key1'],axis=1, inplace=True)
df_merge3.head(2)

#Final File
df_Final_Jul=df_merge3.rename(columns={'User ID_y':'ID','month_y':'month','concat_y':'concat','sessions_y':'sessions'}) 
df_Final_Jul.head()
df_Final_Jul = df_Final_Jul[['ID', 'sessions', 'month', 'concat','User Query', 'Bot Response','Intent Recognized']]
df_Final_Jul.head(2)
df_Final_Jul.shape
```

```python
#filter Month - Dec
data_12 = data[data['month']== 12]
data_12.shape

#Sort Data 
data_Dec = data_12.sort_values(by=['User ID','Time1','Number'], ascending=[1,1,0])
data_Dec = data_Dec[['User ID','Time1','User Query','Intent Recognized','Bot Response','Date_str','month']]
data_Dec.head()

#Define Sessions

data_Dec['Time'] = pd.to_datetime(data_Dec['Time1'])
data_Dec['diffs'] = data_Dec.groupby('User ID')['Time1'].diff()
data_Dec['break'] = data_Dec['diffs'] >= pd.Timedelta('10 min')
data_Dec['sessions'] = data_Dec.groupby('User ID')['break'].cumsum().apply(lambda x: 'session_{}'.format(x+1))
data_Dec.head(2)

#For Joining convert Variable to str

data_Dec['id_str']=data_Dec['User ID'].astype('str')
data_Dec['Date_str2']=data_Dec['Date_str'].astype('str')
data_Dec['Session_str']=data_Dec['sessions'].astype('str')

#Concat all three variables for joining
data_Dec['concat'] = data_Dec['id_str'] + data_Dec['Date_str2']+ data_Dec['Session_str']

# Create Unique Base for Joining
data_Base_Dec=data_Dec[['User ID','Date_str2','concat','month','sessions']]
data_Base_Dec.drop_duplicates(keep='first', inplace=True)
data_Base_Dec.head()
data_Base_Dec.shape

#User Query Transpose
data_Dec_User = data_Dec.groupby(['User ID','sessions','month','concat'])['User Query'].apply(lambda x: ' > '.join(x)).reset_index()
data_Dec_User['User Query'] = data_Dec_User['User Query'].str.lower()
data_Dec_User.head()

#Bot Response Transpose
data_Dec_Bot = data_Dec.groupby(['User ID','sessions','month','concat'])['Bot Response'].apply(lambda x: ' > '.join(x)).reset_index()
data_Dec_Bot['Bot Response'] = data_Dec_Bot['Bot Response'].str.lower()
data_Dec_Bot.head()

#Intent Response Transpose
data_Dec_intent = data_Dec.groupby(['User ID','sessions','month','concat'])['Intent Recognized'].apply(lambda x: ' > '.join(x)).reset_index()
data_Dec_intent['Intent Recognized'] = data_Dec_intent['Intent Recognized'].str.lower()
data_Dec_intent.head()

#Merge1
data_Base_Dec['key1'] = data_Base_Dec.concat.str.strip()
data_Dec_User['key1'] = data_Dec_User.concat.str.strip()
df_merge1 = pd.merge(data_Base_Dec,data_Dec_User, how='left', left_on=['key1'], right_on=['key1'])
df_merge1.head()
df_merge1.columns
df_merge1.drop(['key1','User ID_y', 'sessions_y', 'month_y', 'concat_y'],axis=1, inplace=True)
df_merge1.head(2)


#Merge2
df_merge1['key1'] = df_merge1.concat_x.str.strip()
data_Dec_Bot['key1'] = data_Dec_Bot.concat.str.strip()
df_merge2 = pd.merge(df_merge1,data_Dec_Bot, how='left', left_on=['key1'], right_on=['key1'])
df_merge2.head()
df_merge2.drop(['User ID_x', 'concat_x', 'month_x', 'sessions_x','key1'],axis=1, inplace=True)
df_merge2.head(2)

#Merge3
df_merge2['key1'] = df_merge2.concat.str.strip()
data_Dec_intent['key1'] = data_Dec_intent.concat.str.strip()
df_merge3 = pd.merge(df_merge2,data_Dec_intent, how='left', left_on=['key1'], right_on=['key1'])
df_merge3.head()
df_merge3.drop(['User ID_x', 'sessions_x', 'month_x', 'concat_x','key1'],axis=1, inplace=True)
df_merge3.head(2)

#Final File
df_Final_Dec=df_merge3.rename(columns={'User ID_y':'ID','month_y':'month','concat_y':'concat','sessions_y':'sessions'}) 
df_Final_Dec.head()
df_Final_Dec = df_Final_Dec[['ID', 'sessions', 'month', 'concat','User Query', 'Bot Response','Intent Recognized']]
df_Final_Dec.head(2)
df_Final_Dec.shape
```
