```python
def cut(arr):
    bins = np.empty(arr.shape[0])
    for idx, x in enumerate(arr):
        if (x == 1) :
            bins[idx] = 1
        elif (x == 2):
            bins[idx] = 2
        elif (x >= 3)&(x<=5):
            bins[idx] = 3
        else:
            bins[idx] = 4

    return bins

conversion_dict = {1: '1',
                   2: '2',
                   3: '3-5',
                   4: '>5'}
```

```python
a = cut(df_escalaton['query_count'].to_numpy())

bins = list(map(conversion_dict.get, a))
df_escalaton['query_bin']=bins
```

```python
df_escalaton.groupby(['Month_y','query_bin']).agg({'clientcode':'nunique', 'query_count':'sum'}).reset_index()
```
